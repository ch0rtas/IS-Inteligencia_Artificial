```python
# Instalar las librerías necesarias sin mostrar detalles
!pip install tensorflow nltk rouge-score > /dev/null 2>&1

# Importar las librerías necesarias
import tensorflow as tf
import random
import nltk
from google.colab import files

# Cargar los modelos entrenados
model_sanchez = tf.keras.models.load_model('sanchez.keras')
model_casado = tf.keras.models.load_model('casado.keras')
model_abascal = tf.keras.models.load_model('abascal.h5')

# Cargar los archivos de texto con las intervenciones
with open('intervencionesSanchez.txt', 'r', encoding='utf-8') as file:
    corpus_sanchez = file.readlines()

with open('intervencionesCasado.txt', 'r', encoding='utf-8') as file:
    corpus_casado = file.readlines()

with open('intervencionesAbascal.txt', 'r', encoding='utf-8') as file:
    corpus_abascal = file.readlines()

# Función para generar texto a partir de una entrada con un número aleatorio de palabras entre 90 y 130
def generate_text(model, seed_text, num_words_min=90, num_words_max=130):
    """Genera texto usando el modelo entrenado, con un número aleatorio de palabras entre num_words_min y num_words_max."""
    num_words = random.randint(num_words_min, num_words_max)
    input_text = seed_text
    for _ in range(num_words):
        # Simulación de predicción basada en el corpus por ahora
        next_word = random.choice(" ".join(corpus_sanchez).split())  # Generación simplificada
        input_text += ' ' + next_word
    return input_text

# Función para extraer las últimas 30 palabras de un texto
def get_last_30_words(text):
    """Extrae las últimas 30 palabras de un texto."""
    words = text.split()
    return ' '.join(words[-30:])

# Función para simular un diálogo entre los tres modelos
def generate_dialogue():
    # Inicializar el diálogo con una frase aleatoria del corpus de Sánchez
    phrase_sanchez = random.choice(corpus_sanchez).strip()

    # Primera ronda de interacción
    response_sanchez = generate_text(model_sanchez, phrase_sanchez)
    print(f"--- Sánchez (Respuesta 1) ---\n{response_sanchez}\n")

    phrase_casado_30 = get_last_30_words(response_sanchez)
    response_casado = generate_text(model_casado, phrase_casado_30)
    print(f"--- Casado (Respuesta 1) ---\n{response_casado}\n")

    phrase_abascal_30 = get_last_30_words(response_casado)
    response_abascal = generate_text(model_abascal, phrase_abascal_30)
    print(f"--- Abascal (Respuesta 1) ---\n{response_abascal}\n")

    # Segunda ronda de interacción
    phrase_sanchez_30 = get_last_30_words(response_abascal)
    response_sanchez = generate_text(model_sanchez, phrase_sanchez_30)
    print(f"--- Sánchez (Respuesta 2) ---\n{response_sanchez}\n")

    phrase_casado_30 = get_last_30_words(response_sanchez)
    response_casado = generate_text(model_casado, phrase_casado_30)
    print(f"--- Casado (Respuesta 2) ---\n{response_casado}\n")

    phrase_abascal_30 = get_last_30_words(response_casado)
    response_abascal = generate_text(model_abascal, phrase_abascal_30)
    print(f"--- Abascal (Respuesta 2) ---\n{response_abascal}\n")

    # Tercera ronda de interacción
    phrase_sanchez_30 = get_last_30_words(response_abascal)
    response_sanchez = generate_text(model_sanchez, phrase_sanchez_30)
    print(f"--- Sánchez (Respuesta 3) ---\n{response_sanchez}\n")

    phrase_casado_30 = get_last_30_words(response_sanchez)
    response_casado = generate_text(model_casado, phrase_casado_30)
    print(f"--- Casado (Respuesta 3) ---\n{response_casado}\n")

    phrase_abascal_30 = get_last_30_words(response_casado)
    response_abascal = generate_text(model_abascal, phrase_abascal_30)
    print(f"--- Abascal (Respuesta 3) ---\n{response_abascal}\n")

    # Cuarta ronda de interacción
    phrase_sanchez_30 = get_last_30_words(response_abascal)
    response_sanchez = generate_text(model_sanchez, phrase_sanchez_30)
    print(f"--- Sánchez (Respuesta 4) ---\n{response_sanchez}\n")

    phrase_casado_30 = get_last_30_words(response_sanchez)
    response_casado = generate_text(model_casado, phrase_casado_30)
    print(f"--- Casado (Respuesta 4) ---\n{response_casado}\n")

    phrase_abascal_30 = get_last_30_words(response_casado)
    response_abascal = generate_text(model_abascal, phrase_abascal_30)
    print(f"--- Abascal (Respuesta 4) ---\n{response_abascal}\n")

    # Quinta ronda de interacción
    phrase_sanchez_30 = get_last_30_words(response_abascal)
    response_sanchez = generate_text(model_sanchez, phrase_sanchez_30)
    print(f"--- Sánchez (Respuesta 5) ---\n{response_sanchez}\n")

    phrase_casado_30 = get_last_30_words(response_sanchez)
    response_casado = generate_text(model_casado, phrase_casado_30)
    print(f"--- Casado (Respuesta 5) ---\n{response_casado}\n")

    phrase_abascal_30 = get_last_30_words(response_casado)
    response_abascal = generate_text(model_abascal, phrase_abascal_30)
    print(f"--- Abascal (Respuesta 5) ---\n{response_abascal}\n")

# Ejecutar el diálogo
generate_dialogue()
```
